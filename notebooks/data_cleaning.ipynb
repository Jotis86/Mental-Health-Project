```python
# Check for outliers in the numerical columns with a boxplot
plt.figure(figsize=(15, 8))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title('Outliers Detection in Numerical Columns')
plt.xticks(rotation=45)
plt.show()
```

```python
# Function to detect outliers using the IQR method
def detect_outliers_iqr(df):
    # Select only numeric columns
    df_numeric = df.select_dtypes(include=[np.number])
    
    Q1 = df_numeric.quantile(0.25)
    Q3 = df_numeric.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).sum()
    return outliers

# Detect outliers using the IQR method
outliers_iqr = detect_outliers_iqr(df)
print("Outliers detected using IQR method:")
print(outliers_iqr)
```

```python
# Correlation matrix heatmap including all columns
# Convert categorical columns to numerical using Label Encoding
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
le = LabelEncoder()
for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Calculate the correlation matrix
corr = df_encoded.corr()

# Plot the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Correlation Matrix Analysis

The correlation matrix shows the relationship between different variables in your dataset. Correlation values range between -1 and 1:

- **1** indicates a perfect positive correlation.
- **-1** indicates a perfect negative correlation.
- **0** indicates no correlation.

#### Variables with High Positive Correlation

1. **Physical_Activity_Hours and Support_Systems_Access**:
   - **Correlation**: 0.022075
   - **Interpretation**: There is a slight positive correlation between hours of physical activity and access to support systems. This suggests that people who have access to support systems tend to engage in more physical activity.

2. **Mental_Health_Status and Support_Systems_Access**:
   - **Correlation**: 0.023962
   - **Interpretation**: There is a slight positive correlation between mental health status and access to support systems. This suggests that people with better access to support systems tend to report better mental health status.

#### Variables with High Negative Correlation

1. **Sleep_Hours and Work_Environment_Impact**:
   - **Correlation**: -0.023343
   - **Interpretation**: There is a slight negative correlation between hours of sleep and the impact of the work environment. This suggests that a negative work environment may be associated with fewer hours of sleep.

2. **Online_Support_Usage and Physical_Activity_Hours**:
   - **Correlation**: -0.023109
   - **Interpretation**: There is a slight negative correlation between the use of online support and hours of physical activity. This suggests that people who spend more time using online support tend to engage in less physical activity.

#### Variables with Near-Zero Correlation

1. **Technology_Usage_Hours and Social_Media_Usage_Hours**:
   - **Correlation**: 0.023199
   - **Interpretation**: There is a very low correlation between hours of technology usage and hours of social media usage. This suggests that technology usage is not strongly related to social media usage in this dataset.

2. **Gender and Technology_Usage_Hours**:
   - **Correlation**: -0.000082
   - **Interpretation**: There is practically no correlation between gender and hours of technology usage. This suggests that technology usage is similar across different genders in this dataset.
