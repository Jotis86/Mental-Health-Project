```python
# Import libraries for data analysis and plotting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
df = pd.read_csv('data/data.csv')
```

```python
# Apply set style to seaborn
sns.set_style("whitegrid")
```

```python
# Plot 1: Distribution of the age column
plt.figure(figsize=(12,6))
sns.histplot(df['Age'], bins=30, color='blue', kde=True, palette='muted')
plt.title('Distribution of Age', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
# Save the plot to the reports/figures directory
plt.savefig('reports/figures/age_distribution.png')
```

```python
# Plot 2: Show a range of ages and their counts

# Define the age range bins and corresponding labels
bins = [0, 18, 35, 50, 65, 100]
labels = ['0-17', '18-34', '35-49', '50-64', '65+']

# Create a new column 'Age_Group' with the age ranges
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Plot 2: Show a range of ages and their counts
plt.figure(figsize=(14, 8))
sns.set_style(style="whitegrid")

# Create the histogram
ax = sns.histplot(df['Age_Group'], color='skyblue', kde=False, edgecolor='black')

# Add titles and labels
plt.title('Distribution of Age Groups', fontsize=22, weight='bold')
plt.xlabel('Age Group', fontsize=18)
plt.ylabel('Count', fontsize=18)

# Customize the ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add value labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 10), 
                textcoords='offset points')

# Show the plot
plt.show()

# Save the plot to the reports/figures directory
plt.savefig('reports/figures/Distribution of Age Groups.png')
```

```python
# Plot 3: Pie chart of the age groups

# Define the age range bins and corresponding labels
bins = [18, 35, 50, 65, 100]
labels = ['18-34', '35-49', '50-64', '65+']

# Create a new column 'Age_Group' with the age ranges
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculate the counts for each age group
age_group_counts = df['Age_Group'].value_counts().sort_index()

# Plot: Pie chart of age groups
plt.figure(figsize=(12, 8))
colors = sns.color_palette("pastel")
wedges, texts, autotexts = plt.pie(age_group_counts, autopct='%1.1f%%', startangle=140, colors=colors, textprops=dict(color="w"))

# Add a legend
plt.legend(wedges, labels, title="Age Groups", loc="upper right", bbox_to_anchor=(1, 1))

# Add a title
plt.title('Distribution of Age Groups', fontsize=20, weight='bold')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()

# Save the plot to the reports/figures directory
plt.savefig('reports/figures/Distribution of Age Groups_2.png')
```

```python
# Plot 4: Countplot of the gender column
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title('Gender Distribution', fontsize=20, weight='bold')
plt.xlabel('Gender', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
# Save the plot to the reports/figures directory
plt.savefig('reports/figures/Gender Distribution.png')
```

```python
# Plot 5: Pie chart of the gender column
# Calculate the counts for each gender
gender_counts = df['Gender'].value_counts()

# Plot: Pie chart of gender
plt.figure(figsize=(8, 8))
colors = sns.color_palette("pastel")
wedges, texts, autotexts = plt.pie(gender_counts, autopct='%1.1f%%', startangle=140, colors=colors, textprops=dict(color="w"))

# Add a legend
plt.legend(wedges, gender_counts.index, title="Gender", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))

# Add a title
plt.title('Gender Distribution', fontsize=20, weight='bold')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()

# Save the plot to the reports/figures directory
plt.savefig('reports/figures/Gender Distribution_2.png')
```
